**RGB Image Conversion**

```python
import cv2
import matplotlib.pyplot as plt

# Load an RGB image
image_path = 'lighthouse.jpg'  
image = cv2.imread(image_path)

# loads images in BGR format, convert it to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image (BGR)')

plt.subplot(1, 2, 2)
plt.imshow(image_rgb)
plt.title('RGB Image')

plt.show()
```

**RGB to CMY**

```python
from PIL import Image
import matplotlib.pyplot as plt

def rgb_to_cmy(rgb_image_path, cmy_image_path):
    rgb_image = Image.open(rgb_image_path)

    # Create a new blank CMY image with the same size
    cmy_image = Image.new('RGB', rgb_image.size)

    for x in range(rgb_image.width):
        for y in range(rgb_image.height):
            # Get the RGB values of the pixel
            r, g, b = rgb_image.getpixel((x, y))

            # Convert RGB to CMY
            cyan = 1 - r / 255.0
            magenta = 1 - g / 255.0
            yellow = 1 - b / 255.0

            # Scale CMY values to the range [0, 255] for creating a new pixel
            c = int(cyan * 255)
            m = int(magenta * 255)
            y = int(yellow * 255)

            # Set the new pixel in the CMY image
            cmy_image.putpixel((x, y), (c, m, y))

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb_image)
    plt.title('Original RGB Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cmy_image)
    plt.title('Converted CMY Image')
    plt.axis('off')

    cmy_image.save(cmy_image_path)

    plt.show()

rgb_image_path = 'lighthouse.jpg'
cmy_image_path = 'lighthouse_cmy.jpg'
rgb_to_cmy(rgb_image_path, cmy_image_path)
```

**RGB to CMYK**

```python
from PIL import Image
import matplotlib.pyplot as plt

def rgb_to_cmyk(rgb_image_path, cmyk_image_path):
    rgb_image = Image.open(rgb_image_path)

    # Create a new blank CMYK image with the same size
    cmyk_image = Image.new('CMYK', rgb_image.size)

    # Loop through each pixel in the RGB image
    for x in range(rgb_image.width):
        for y in range(rgb_image.height):
            # Get the RGB values of the pixel
            r, g, b = rgb_image.getpixel((x, y))

            # Convert RGB to CMY
            cyan = 1 - r / 255.0
            magenta = 1 - g / 255.0
            yellow = 1 - b / 255.0

            # Calculate Black (Key)
            black = min(cyan, magenta, yellow)

            # Scale CMYK values to the range [0, 255] for creating a new pixel
            c = int(cyan * 255)
            m = int(magenta * 255)
            y = int(yellow * 255)
            k = int(black * 255)

            # Set the new pixel in the CMYK image
            cmyk_image.putpixel((x, y), (c, m, y, k))

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb_image)
    plt.title('Original RGB Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cmyk_image)
    plt.title('Converted CMYK Image')
    plt.axis('off')

    cmyk_image.save(cmyk_image_path)

    plt.show()

rgb_image_path = 'lighthouse.jpg'
cmyk_image_path = 'lighthouse_cmyk.jpg'
rgb_to_cmyk(rgb_image_path, cmyk_image_path)
```

**RGB to HSV**

```python
from PIL import Image
import matplotlib.pyplot as plt
import colorsys

def rgb_to_hsv(rgb_image_path, hsv_image_path):
    rgb_image = Image.open(rgb_image_path)

    hsv_image = Image.new('HSV', rgb_image.size)

    for x in range(rgb_image.width):
        for y in range(rgb_image.height):
            r, g, b = rgb_image.getpixel((x, y))

            # Convert RGB to HSV
            h, s, v = colorsys.rgb_to_hsv(r / 255.0, g / 255.0, b / 255.0)

            # Scale HSV values to the range [0, 255] for creating a new pixel
            h = int(h * 255)
            s = int(s * 255)
            v = int(v * 255)

            # Set the new pixel in the HSV image
            hsv_image.putpixel((x, y), (h, s, v))

    # Convert HSV image to RGB before saving as JPEG
    rgb_hsv_image = hsv_image.convert('RGB')

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb_image)
    plt.title('Original RGB Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rgb_hsv_image)
    plt.title('Converted HSV Image')
    plt.axis('off')

    # Save the resulting HSV image as JPEG
    rgb_hsv_image.save(hsv_image_path)

    plt.show()


rgb_image_path = 'lighthouse.jpg'
hsv_image_path = 'lighthouse_hsv.jpg'
rgb_to_hsv(rgb_image_path, hsv_image_path)
```

**RGB to HSL**

```python
from PIL import Image
import matplotlib.pyplot as plt
import colorsys

def rgb_to_hsl(rgb_image_path, hsl_image_path):
    rgb_image = Image.open(rgb_image_path)

    # Create a new blank HSL image with the same size
    hsl_image = Image.new('RGB', rgb_image.size)

    # Loop through each pixel in the RGB image
    for x in range(rgb_image.width):
        for y in range(rgb_image.height):
            # Get the RGB values of the pixel
            r, g, b = rgb_image.getpixel((x, y))

            # Convert RGB to HSL
            h, l, s = colorsys.rgb_to_hls(r / 255.0, g / 255.0, b / 255.0)

            # Scale HSL values to the range [0, 255] for creating a new pixel
            h = int(h * 255)
            l = int(l * 255)
            s = int(s * 255)

            # Set the new pixel in the HSL image
            hsl_image.putpixel((x, y), (h, l, s))

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb_image)
    plt.title('Original RGB Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(hsl_image)
    plt.title('Converted HSL Image')
    plt.axis('off')

    hsl_image.save(hsl_image_path)

    plt.show()

rgb_image_path = 'lighthouse.jpg'
hsl_image_path = 'lighthouse_hsl.jpg'
rgb_to_hsl(rgb_image_path, hsl_image_path)
```

**RGB to YUV**

```python
from PIL import Image
import matplotlib.pyplot as plt

def rgb_to_yuv(rgb_image_path, yuv_image_path):
    # Open the RGB image
    rgb_image = Image.open(rgb_image_path)

    # Create a new blank YUV image with the same size
    yuv_image = Image.new('YCbCr', rgb_image.size)

    # Loop through each pixel in the RGB image
    for x in range(rgb_image.width):
        for y in range(rgb_image.height):
            # Get the RGB values of the pixel
            r, g, b = rgb_image.getpixel((x, y))

            # Convert RGB to YUV
            y = int(0.299 * r + 0.587 * g + 0.114 * b)
            u = int(-0.14713 * r - 0.288862 * g + 0.436 * b + 128)
            v = int(0.615 * r - 0.51498 * g - 0.10001 * b + 128)

            # Set the new pixel in the YUV image
            yuv_image.putpixel((x, y), (y, u, v))

    # Display the original RGB image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb_image)
    plt.title('Original RGB Image')
    plt.axis('off')

    # Display the resulting YUV image
    plt.subplot(1, 2, 2)
    plt.imshow(yuv_image)
    plt.title('Converted YUV Image')
    plt.axis('off')

    # Save the resulting YUV image
    yuv_image.save(yuv_image_path)

    # Show the plots
    plt.show()


rgb_image_path = 'lighthouse.jpg'
yuv_image_path = 'lighthouse_yuv.jpg'
rgb_to_yuv(rgb_image_path, yuv_image_path)
```
