**Brightness**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('lighthouse.jpg')

# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Increase brightness by 50 (specify dtype=np.float64)
brightened_image = cv2.add(image, np.array([50], dtype=np.float64))

# Display brightened image
plt.subplot(1, 2, 2)
plt.title('Brightened Image')
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Contrast**

```python
# Assuming the 'brightened_image' is the previously processed image
# Display contrasted image
plt.subplot(1, 2, 1)
plt.title('Brightened Image')
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Increase contrast by scaling pixel values
contrast_factor = 1.5  # Adjust as needed
contrasted_image = cv2.multiply(brightened_image, contrast_factor)

# Display contrasted image
plt.subplot(1, 2, 2)
plt.title('Contrasted Image')
plt.imshow(cv2.cvtColor(contrasted_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Sharpening**

```python
# Assuming the 'contrasted_image' is the previously processed image
# Display sharpened image
plt.subplot(1, 2, 1)
plt.title('Brightened Image')
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply a simple sharpening kernel
sharpening_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv2.filter2D(brightened_image, -1, sharpening_kernel)

# Display sharpened image
plt.subplot(1, 2, 2)
plt.title('Sharpened Image')
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Blur**

```python
# Assuming the 'sharpened_image' is the previously processed image
# Display original image
plt.subplot(1, 2, 1)
plt.title('Sharpened Image')
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(sharpened_image, (5, 5), 0)

# Display blurred image
plt.subplot(1, 2, 2)
plt.title('Blurred Image')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Smooth**

```python
# Assuming the 'blurred_image' is the previously processed image
# Display original image
plt.subplot(1, 2, 1)
plt.title('Blurred Image')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply bilateral filter for smoothing while preserving edges
smoothed_image = cv2.bilateralFilter(blurred_image, 9, 75, 75)

# Display smoothed image
plt.subplot(1, 2, 2)
plt.title('Smoothed Image')
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Illumination**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('lighthouse.jpg')

# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Adjust illumination (gamma correction)
gamma_value = 1.5  # Adjust as needed
illuminated_image = np.power(image / 255.0, gamma_value) * 255.0
illuminated_image = np.clip(illuminated_image, 0, 255).astype(np.uint8)

# Display illuminated image
plt.subplot(1, 2, 2)
plt.title('Illuminated Image')
plt.imshow(cv2.cvtColor(illuminated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Defocus**

```python
# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Simulate defocused image using Gaussian blur
defocus_amount = 5  # Adjust as needed
defocused_image = cv2.GaussianBlur(image, (defocus_amount, defocus_amount), 0)

# Display defocused image
plt.subplot(1, 2, 2)
plt.title('Defocused Image')
plt.imshow(cv2.cvtColor(defocused_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```

**Focal Length**

```python

# Display original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Simulate change in focal length (scaling)
focal_length_factor = 1.5  # Adjust as needed
rescaled_image = cv2.resize(image, None, fx=focal_length_factor, fy=focal_length_factor, interpolation=cv2.INTER_LINEAR)

# Display image with simulated change in focal length
plt.subplot(1, 2, 2)
plt.title('Focal Length Change')
plt.imshow(cv2.cvtColor(rescaled_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the plot
plt.show()
```
